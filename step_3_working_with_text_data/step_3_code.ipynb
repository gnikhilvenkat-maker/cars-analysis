```python
import pandas as pd
import os
```

```python
df = pd.read_csv("../step_1_reading_and_preprocessing/tables/cleaned_data.csv")
df.head()
```

```python
models_per_company = df.groupby("Company Names")["Cars Names"].nunique().reset_index()
models_per_company = models_per_company.sort_values("Cars Names", ascending=False)
```

```python
models_per_company.to_csv("../step_3_working_with_text_data/tables/models_per_company.csv", index=False)
models_per_company.head()
```

```python
electric_companies = df[df["Fuel_Type_clean"]=="Electric"].groupby("Company Names").size().reset_index(name="count")
electric_companies = electric_companies.sort_values("count", ascending=False).head(5)
```

```python
electric_companies.to_csv("../step_3_working_with_text_data/tables/top5_companies_electric.csv", index=False)
electric_companies
```
