```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../step_1_reading_and_preprocessing/tables/cleaned_data.csv")
```

```python
manufacturer = "Toyota"
sub = df[df["Company Names"].str.lower()==manufacturer.lower()]
plt.hist(sub["Price_clean"].dropna(), bins=20)
plt.title(f"Price Distribution - {manufacturer}")
plt.xlabel("Price")
plt.ylabel("Count")
plt.savefig("plots/price_histogram.png")
plt.show()
```

```python
top5 = df.nlargest(5,"HorsePower_clean")
plt.scatter(top5["HorsePower_clean"], top5["Total_Speed_clean"])
for i,row in top5.iterrows():
    plt.text(row["HorsePower_clean"], row["Total_Speed_clean"], row["Cars Names"], fontsize=8)
plt.xlabel("HorsePower")
plt.ylabel("Total Speed")
plt.title("Speed vs Power (Top 5 Cars)")
plt.savefig("plots/scatter_speed_power.png")
plt.show()
```

```python
top10 = df.groupby("Company Names")["Cars Names"].nunique().reset_index()
top10 = top10.sort_values("Cars Names", ascending=False).head(10)
plt.bar(top10["Company Names"], top10["Cars Names"])
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Companies by Models")
plt.ylabel("Number of Models")
plt.savefig("plots/top10_companies.png")
plt.show()
```
