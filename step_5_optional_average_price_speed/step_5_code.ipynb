```python
import pandas as pd
import os
```

```python
df = pd.read_csv("../step_1_reading_and_preprocessing/tables/cleaned_data.csv")
```

```python
avg_price = df["Price_clean"].mean()
low, high = avg_price*0.9, avg_price*1.1
mask = (df["Price_clean"]>=low)&(df["Price_clean"]<=high)
top10_speed = df[mask].nlargest(10, "Total_Speed_clean")[['Company Names','Cars Names','Price_clean','Total_Speed_clean']]
top10_speed.to_csv("tables/top10_fast_in_avg_price_window.csv", index=False)
top10_speed
```
