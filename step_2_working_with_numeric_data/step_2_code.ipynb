```python
import pandas as pd
import os
```

```python
df = pd.read_csv("../step_1_reading_and_preprocessing/tables/cleaned_data.csv")
df.head()
```

```python
avg_price = df.groupby("Company Names")["Price_clean"].mean().reset_index()
```

```python
avg_price.to_csv("../step_2_working_with_numeric_data/tables/avg_price_by_company.csv", index=False)
avg_price.head()
```

```python
top5_power = df.nlargest(5, "HorsePower_clean")[["Company Names","Cars Names","HorsePower_clean"]]
top5_power.to_csv("../step_2_working_with_numeric_data/tables/top5_powerful_cars.csv", index=False)
top5_power
```

```python
perf = df.groupby("Fuel_Type_clean")["Performance_clean"].mean().reset_index()
perf.to_csv("../step_2_working_with_numeric_data/tables/performance_by_fuel.csv", index=False)
perf
```
